### Goal:

Create a dataframe that holds:

- name of the renter, 
- whether their lease of renewed or terminated,
- the name of the source file.

````python
import glob
import pandas as pd
````

## Glob all target documents

````python
## pull into only txt files into notebook
target_files = sorted(glob.glob("my_downloads/*.txt"))
target_files
````

````python
for target_file in target_files:
    with open(target_file, "r") as my_text:
        print(f"Source file: {target_file}")
        all_text = my_text.readlines()
        print(f"List: {all_text}")
        print(f"Client: {all_text[0]}")
        
````

````python
decisions_list = []
for target_file in target_files:
    with open(target_file, "r") as my_text:
        all_text = my_text.readlines()
        client = all_text[0].replace("Client: ", "").rstrip()
        print(client)
        decision = all_text[2]
        print(decision)
        if "renew" in decision:
            decision = "renew"
        else: 
            decision = "terminate"  
        print(decision)
        decisions_list.append({"client": client,
                               "decision": decision,
                              "source_file": target_file})
        
````

````python
decisions_list
````

````python
df = pd.DataFrame(decisions_list)
df
````
