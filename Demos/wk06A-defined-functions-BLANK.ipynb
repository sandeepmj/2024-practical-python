## ```Docstrings```

As our functions get more complicated, it gets harder to remember what the parameters are and what they actually do. ```Docstrings``` help us and our colleagues understand what a function does; what parameters you need; and what order they need to be in.

Info to include in a ```docstring```:

* **A brief summary of what the function does.**
* **A description of each ```parameter``` and their expected data types.**
* **The `Returns` section specifies what value (if any) the function
returns.**
* The `Notes` section provides additional context or guidelines
for using the function.
* The `Example usage` section shows how to use the function in a simple
example.
